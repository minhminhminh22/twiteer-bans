```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('trainset.csv')
```

```python
# Reading from the file
df.head()
```

```python
# statisctics of the dataset
df.describe()
```

```python
# plotting the total number of tweets a tweet has
df.hist('totaltweets')
```

```python
#to see if the tweets mention abortion
df['has abortion'] = df['text'].apply(lambda x: ('abortion' in x.lower()))
```

```python
#counting the number of tweets that mentions abortion
df['has abortion'].value_counts()
```

```python
#showing the tweets that mentions abortion
df['text'].values
```

```python
#create a search condition "has Supreme Court"
df['has Supreme Court'] = df['text'].apply(lambda x: ('Supreme Court' in x))
```

```python
#count the tweets that has and not have "Supreme Court"
df['has Supreme Court'].value_counts()
```

```python
#modify the search term "has Supreme Court" 
df['has Supreme Court case-insensitive'] = df['text'].apply(lambda x: ('supreme court' in x.lower()))
```

```python
#re-count with the modified search term
df['has Supreme Court case-insensitive'].value_counts()
```

```python
#list of tweets testing with the conditions of search terms
df[(df['has Supreme Court']==False)&(df['has Supreme Court case-insensitive']==True)]
```

```python
#number of retweets a tweet that mentions the Supreme Court has
number_of_retweets = df[df['has Supreme Court case-insensitive'] == True]
```

```python
#display the tweets that match the requirements above
number_of_retweets
```

```python
#display the number of retweets they have
number_of_retweets[['retweetcount', 'has Supreme Court case-insensitive']]
```

```python
# number of retweets a tweet that do not mentions the Supreme Court has
retweets = df[df['has Supreme Court case-insensitive']==False]
```

```python
# display the number of retweets they have
retweets[['retweetcount', 'has Supreme Court case-insensitive']]
```

```python
#write data to another file
false_SCOTUS= retweets[['retweetcount', 'has Supreme Court case-insensitive']]
false_SCOTUS.to_csv('no_SCOTUS.csv')
```
